```python
import numpy as np
import xarray as xr

# Rango de deltaT y deltaP
deltaT_values = np.arange(-5, 6, 1)
deltaP_values = np.arange(-10, 11, 1)
days = np.arange(0, 101, 1) #estimate with the length of the simulation period

# Simulación ficticia del rendimiento (puedes reemplazar con WOFOST real)
def simulate_wofost(deltaT, deltaP, days):
    return np.sin(0.1 * days + deltaT) + np.cos(0.1 * days + deltaP)

# Crear array de datos
data = np.zeros((len(deltaT_values), len(deltaP_values), len(days)))
for i, dt in enumerate(deltaT_values):
    for j, dp in enumerate(deltaP_values):
        data[i, j, :] = simulate_wofost(dt, dp, days)

# Crear xarray
yield_data = xr.DataArray(
    data,
    coords={'deltaT': deltaT_values, 'deltaP': deltaP_values, 'time': days},
    dims=['deltaT', 'deltaP', 'time'],
    name='yield'
)

print(yield_data)
yield_data.mean(dim='time').plot()  # Ejemplo de visualización del rendimiento promedio
```
